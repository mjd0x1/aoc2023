```python
import numpy as np
import pandas as pd

#open text file
with open('input9.txt') as f:
    data = f.readlines()

history = []
for x in data:
    history.append([int(t) for t in x.strip().split(' ')])

sum_s = 0
for h  in history:
    z = np.array(h)
    r = []
    while not np.all(np.diff(z) == 0):
        r.append(np.diff(z))
        z = np.diff(z)
    print("*",r)
    print("**", h)
    for i,t in enumerate(r):
        sum_s+=t[0]*((-1)**(i+1))
    


for h in history:
    sum_s += h[0]    

sum_s
```
