```python
import numpy as np
import pandas as pd

#open text file
with open('input12.txt') as f:
    data = f.readlines()
patterns = []
counts = []
for x in data:
    patterns.append(x.strip().split(" ")[0])
    ct = x.strip().split(" ")[1]
    counts.append(list(map(lambda x: int(x), ct.split(","))))
counts,patterns
```

```python


def get_count(s):
    return [len(t)  for t in [x for x in s.split(".") if "#" in x]]

def process(s,depth): 

    all_counts = []
    def process_(s,depth):
        
        if depth ==len(s):
            all_counts.append(s)
            return
        else:
            if s[depth] == "?":
                temp = list(s)
                temp[depth]="."
                process_("".join(temp),depth+1)
                temp = list(s)
                temp[depth]="#"
                process_("".join(temp),depth+1)
            else:
                process_(s,depth+1)
    process_(s,depth)
    return all_counts

def process_all(patterns,counts):
    output = []
    for i,s in enumerate(patterns):
        all_counts = process(s,0)
        sum_s = 0
        for c in all_counts:
            sum_s += (get_count(c)==counts[i])

        sum_s
        output.append(sum_s)
    return np.sum(output)

process_all(patterns,counts)
```
