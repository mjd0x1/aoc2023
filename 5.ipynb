```python
import numpy as np
import pandas as pd

#open text file
with open('input5.txt') as f:
    data = f.readlines()


seeds = list(map(lambda x: int(x), data[0].split(":")[1].strip().split(" ")))


str_maps = [("seed-to-soil","soil-to-fertilizer"),
            ("soil-to-fertilizer","fertilizer-to-water"),
            ("fertilizer-to-water","water-to-light"),
            ("water-to-light","light-to-temperature"),
            ("light-to-temperature","temperature-to-humidity"),
            ("temperature-to-humidity","humidity-to-location"),
            ("humidity-to-location" , "-1")
    
]

p= np.array(seeds).reshape(int(len(seeds)/2),2).tolist()
#new_seeds = []
#for x in p:
#    new_seeds+= list(range(x[0],x[0]+10000))

def generate_map(strid, strpost):
    id1 = np.where(np.array(data) == strid + " map:\n")[0][0]
    if strpost != "-1":
        id2 = np.where(np.array(data) == strpost + " map:\n")[0][0] 
    else:
        id2 = len(data)
    
    map_d =[]
    
    for x in data[id1+1:id2-1]:
        map_d.append(list(map(lambda t : int(t), x.strip().split(" "))))
    map_d
    return map_d

dict_maps = {}
for i,x in enumerate(str_maps):
    dict_maps[x[0]] = generate_map(x[0],x[1])



def get_map_val(point_to_map,map_ , strid):
    
    map_d =map_[strid]
 
    map_val =0 
    for p in map_d:
        dest,source,r_range = p
        if point_to_map >= source and point_to_map <= source+r_range-1:
           
            map_val = dest - source
           
    output = point_to_map
    if map_val != 0:
        output = point_to_map+map_val

    return output

p = [[3319921504, 153335682]]
vals = []
for pp in p:
   
    new_seeds = np.arange(3352513254-1000,3352513254+1000,1)
    for s in new_seeds:
        val = s
        for i,x in enumerate(str_maps):
            val = get_map_val(val,dict_maps,x[0])
           
        vals.append(val)
        

    print(f"{pp[0]} value {np.min(vals)}")

 #439018996 
 #225784067
 #224784068
 #215784167
 #108957168
 #108956486 tried
 #108956286
 #108956236
 #108956227

vals

 
```
