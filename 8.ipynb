```python
import numpy as np
import pandas as pd

#open text file
with open('input8.txt') as f:
    data = f.readlines()

data

state = {}
directions  = list(data[0].strip())
for x in data[2:]:
    key = x.split("=")[0].strip()
    left = x.split("=")[1].strip().split(",")[0][1:4]
    right = x.split("=")[1].strip().split(",")[1][1:4]
    state[key] = {"L":left, "R":right}

state
```

```python
states = ['VNA', 'AAA', 'DPA', 'JPA', 'DBA', 'QJA']
#states = ['VNA', 'AAA', 'JPA']
states= states[:2]
#new_state =""
ct = 0

while not("".join([x[-1] for x in states]) == "Z"*len(states)):
    for x in directions:
        for i,s in enumerate(states):
            states[i]=state[states[i]][x]
    ct+=1

ct *len(directions)
```
