```python
import numpy as np
import pandas as pd

#open text file
with open('input15.txt') as f:
    data = f.readlines()

data = data[0].split(',')
def get_value(x):
    current_value = 0
    for c in x:
        current_value+= ord(c)
        current_value *=17
        current_value = current_value % 256

    return current_value

np.sum([get_value(x) for x in data])
```
